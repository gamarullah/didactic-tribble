## Data Sets
The **Indonesian Family Life Survey** (IFLS) is an on-going longitudinal survey in Indonesia conduct by RAND Corporation. The sample is representative of about 83% of the Indonesian population and contains over 30,000 individuals living in 13 of the 27 provinces in the country.

source:
https://www.rand.org/well-being/social-and-behavioral-policy/data/FLS/IFLS.html

For this purpose, I use IFLS wave 5 (2014)

#### Book K (bk)
Respondent is a HH Member 18 Years or Older who is Knowledgeable About Characteristics of Household Members

```python
# Section AR
# contains the characteristics of samples. it also contains a unique ID to help merge the data set

# year 2014
bk_ar1_14 = pd.read_stata(r'C:\Users\amaru\Documents\Datasets\IFLS\IFLS5\Household\bk_ar1.dta')
```

```python
bk_ar1_14.head()
```

#### Book 3A (b3a)
Respondent is an adult 15 years or older

```python
# Section SW (Subjective Well-Being)
# The main variable that I needed is sw00, it contains the categorical data of life satisfaction

# year 2014
b3a_sw_14 = pd.read_stata(r'C:\Users\amaru\Documents\Datasets\IFLS\IFLS5\Household\b3a_sw.dta')
```

```python
b3a_sw_14.head()
```

## Data Cleaning & Processing

#### Individual Characteristics

```python
samples_14 = bk_ar1_14[['pidlink', 'ar09', 'ar07']]
samples_14 = samples_14.rename(columns={'ar09' : 'age', 'ar07' : 'sex'})  # renaming variables

# converting sex variables
sex = []
for x in samples_14['sex']:
    if x == 3:
        sex.append('Female')
    elif x == 1:
        sex.append('Male')
    else:
        sex.append('Unknown')
samples_14['sex'] = sex
        
# in context of age, value of 998 means "don't know" and 999 means either missing or refuse to answer
# althought not much, I prefer it to be dropped
samples_14.drop(samples_14[samples_14['age'] >= 998].index, inplace=True)

# next let's create an age group
age_group = []
for x in samples_14['age']:
    if x >= 15 and x < 25:
        age_group.append('15-24')
    elif x >= 25 and x < 35:
        age_group.append('25-34')
    elif x >= 35 and x < 45:
        age_group.append('35-44')
    elif x >= 45 and x < 55:
        age_group.append('45-54')
    elif x >= 55 and x < 65:
        age_group.append('55-64')
    elif x >= 65:
        age_group.append('65<')
    else:
        age_group.append('Unknown')

samples_14['age group'] = age_group


samples_14 = samples_14.fillna('Undefined') # Redefine any NaN as 'Undefined'

samples_14.info()
samples_14.head()
```

```python
# Duplicate Checker
# some variables from the data set are known to cause duplication on unique ID
def dupli_checker(data, col):             # data = data (table), col = column
    dupli = data[data.duplicated(col)]
    
    print('There are ', len(dupli), 'duplicated rows(s) out of ', len(data))
    
    if len(dupli) != 0:
        print('Dropping....')
        
        data.drop_duplicates(subset=col, keep='first', inplace=True)
        print('There are ', len(samples_14), ' row(s) left')
    
    else:
        print('No action required')
```

```python
# Duplicates check
dupli_checker(samples_14, "pidlink")
```

#### Life Satisfaction

```python
satisfied_14 = b3a_sw_14[['pidlink', 'sw00']]
satisfied_14 = satisfied_14.rename(columns={'sw00':'satisfaction'})

satisfied_14.info()
satisfied_14.head()
```

```python
# Duplicates check
dupli_checker(satisfied_14, 'pidlink')
```

#### Merging

```python
wb_14 = samples_14.merge(satisfied_14, on='pidlink') #
print('There are', len(wb_14), 'matching rows')

wb_14.head()
```

```python
# There are a few minority in unknown sex and unknown age group. So I decide to drop them
wb_14.drop(wb_14[wb_14['sex'] == 'Unknown'].index, inplace=True)
wb_14.drop(wb_14[wb_14['age group'] == 'Unknown'].index, inplace=True)

print(wb_14['sex'].value_counts())
print('*------------------------------------------------*')
print(wb_14['age group'].value_counts())
```

#### Visualization

Let's do a little viz

```python
age_label = tuple(np.unique(wb_14['age group']))
sex_label = tuple(np.unique(wb_14['sex']))

age_fracs = list(wb_14['age group'].value_counts())
sex_fracs = list(wb_14['sex'].value_counts())

# Make figure and axes
fig, axs = plt.subplots(1,2)


# A standard pie plot
axs[0].pie(age_fracs, labels=age_label, autopct='%1.1f%%', shadow=True)
axs[1].pie(sex_fracs, labels=sex_label, autopct='%1.1f%%', shadow=True)

plt.show()
```
